```python
dis = pickle.load(open('dis.pkl', 'rb'))
```

```python
EXTS = {'mp3', 'flac', 'm3u', 'm4a', 'wma'}
```

```python
exts = sorted({n.split('.')[-1].lower() for n in dis})
', '.join(exts)
```

```python
[n for n in dis if n.split('.')[-1].lower() == 'wpl']
```

```python
music = list(filter(lambda n: n.split('.')[-1].lower() in EXTS, dis))
```

```python
len(music)
```

```python
{len(m.split('/')) for m in music}
```

```python
folders = 4
```

```python
[m for m in music if len(m.split('/')) == folders]
```

```python
sorted({x.split('/')[0] for x in [m for m in music if len(m.split('/')) == folders]})
```

```python
from collections import defaultdict

fourth = defaultdict(list)
fourall = defaultdict(list)
for p in [m for m in music if len(m.split('/')) == 4]:
    fourth[p.split('/')[0]].append(p.split('/')[2])
    fourall[p.split('/')[0]].append(p)    
```

```python
for band, extra in fourth.items():
    print(band)
    print(set(extra))
    print('-' * 100)
```
